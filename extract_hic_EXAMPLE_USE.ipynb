# Raw Hi-C map

```python
path = "./example_data/GSE63525_GM12878_diploid_paternal.hic"
normalization = 'VC'
chrom1 = "1"
res = 50000
array = extract_hic(path, res, chrom1, chrom1, normalization)

start, end = 305, 450
hic_orig = array[start:end, start:end]
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.imshow(hic_orig, cmap='hot', vmax=100)
plt.show()
```
